```python
def fact(n): # i know this is the dumb version, but it doesnt matter
    assert(type(n) == int and 0 <= n)
    if(n == 1 or n == 0):
        return 1
    else:
        return n * fact(n-1)
```

```python
cos_coefficients = [FXnum(val=(((-1)**(k))/fact(2*k)),family=FXfamily(n_bits=56, n_intbits=8)).scaledval for k in range(0,7)]

def inttaylorcos(x, coefficients):
    # index for kth term
    k = 0
    # intermediate variable
    temp = 0
    # result take 0th term and add rest
    cosx = coefficients[k]
    # base monomial that will alway be raised to power k
    xsquare = (x*x) >> 56


    k = 1
    x2k = xsquare
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    
    # have to shift back to force size
    k = 2
    x2k = (x2k * xsquare) >> 56
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    k = 3
    x2k = (x2k * xsquare) >> 56
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    k = 4
    x2k = (x2k * xsquare) >> 56
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    k = 5
    x2k = (x2k * xsquare) >> 56
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    k = 6
    x2k = (x2k * xsquare) >> 56
    temp = (x2k * coefficients[k]) >> 56
    cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    #k = 7
    #x2k = (x2k * xsquare) >> 56
    #temp = (x2k * coefficients[k]) >> 56
    #cosx = cosx + temp 
    #print(xminusak)
    #print(lnx)
    
    #k = 8
    #x2k = (x2k * xsquare) >> 56
    #temp = (x2k * coefficients[k]) >> 56
    #cosx = cosx + temp
    #print(xminusak)
    #print(lnx)
    
    #k = 9
    #x2k = (x2k * xsquare) >> 56
    #temp = (x2k * coefficients[k]) >> 56
    #cosx = cosx + temp 
    #print(xminusak)
    #print(lnx)
    
    #k = 10
    #x2k = (x2k * xsquare) >> 56
    #temp = (x2k * coefficients[k]) >> 56
    #cosx = cosx + temp
    #print(xminusak)

    #print(xminusak)
    #print("int representation x: %d, lnx: %d" % (x, lnx))
    #print("fraction representation x: %f, lnx: %f" % (x/(1<<56), lnx/(1<<56)))
    #print("same x but from python math library x: %f, lnx: %f" % (x/(1<<56), math.log(x/(1<<56))))
    return [str(x), str(x/(1<<56)), str(cosx/(1<<56)), str(math.cos(x/(1<<56)))]

random.seed(0)

l = FXnum(-math.pi,FXfamily(n_bits=56, n_intbits=8))
u = FXnum(math.pi,FXfamily(n_bits=56, n_intbits=8))
print("(%s, %s) -> (%d, %d)" % (l.toDecimalString(), u.toDecimalString(),l.scaledval,u.scaledval))
pts = [inttaylorcos(random.randint(l.scaledval,u.scaledval), cos_coefficients) for i in range(1000)]

for x in pts:
    print(x)

df = pd.DataFrame(pts)
df.columns = ["RAND","DEC","PY_TAYLOR_COS","PY_MATH_COS"]

df.to_csv("./cosTest.csv")

coef_df = pd.DataFrame(cos_coefficients)
coef_df.columns = ["COS_COEF"]
coef_df.to_csv("./cosCoef.csv")
```
